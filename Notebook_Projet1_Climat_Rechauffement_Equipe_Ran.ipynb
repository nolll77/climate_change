```
import numpy as np
import pandas as pd
# csv trouver sur le site : 
# https://www.data.gouv.fr/fr/datasets/cait-country-greenhouse-gas-emissions-data/
# Création du dataframe sur les GreenHouse Gaz (Gaz à effet de serre)
df_GHG = pd.read_csv('CAIT Country GHG Emissions.csv',skiprows= 2)
df_GHG.head()
```

Metric tons of carbon dioxide equivalent (MTCO2e) :  
En francais "l'équivalent CO2" est, pour un gaz à effet de serre (GES), la quantité de CO2 qui provoquerait le même forçage radiatif que ce gaz, c'est-à-dire qui aurait la même capacité à retenir le rayonnement solaire.  
Il est exprimé en appliquant un facteur de conversion, le potentiel de réchauffement global (PRG), qui dépend de la longueur de la période considérée.

```
#On met la colonne Year en Index
df_GHG = df_GHG.groupby(['Year']).mean()
#On modifie notre dataframe en sélectionnant que les 4 premières colonnes 
df_GHG = df_GHG[['Total GHG Emissions Including Land-Use Change and Forestry (MtCO?e?)',
                            'Total CO2 (including Land-Use Change and Forestry) (MtCO2)',
                            'Total CH4 (including Land-Use Change and Forestry) (MtCO2e)',
                            'Total N2O (including Land-Use Change and Forestry) (MtCO2e)']]
df_GHG.head()
```

```
# On modifie le titre de la la 1ere colonne avec  MtCO?e? qui devient MtCO2e
df_GHG.columns = ['Total GHG Emissions Including Land-Use Change and Forestry (MtCO2e)',
                            'Total CO2 (including Land-Use Change and Forestry) (MtCO2)',
                            'Total CH4 (including Land-Use Change and Forestry) (MtCO2e)',
                            'Total N2O (including Land-Use Change and Forestry) (MtCO2e)']
df_GHG.head()                           
```

```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(color_codes=True)
#Création du dataframe
df_t = pd.read_csv("/content/GLB.Ts+dSST.csv", skiprows=1)
```

```
# Créer un nouveau dataframe avec un index pour chaque mois
# Créez d'abord la plage de dates
date_rng = pd.date_range(start='1/1/1880', end='1/03/2019', freq='M')

type(date_rng[0])
### pandas._libs.tslibs.timestamps.Timestamp

# Créez ensuite le DataFrame vide, que nous allons remplir en utilisant les données réelles
t = pd.DataFrame(date_rng, columns=['date'])

# Créer une colonne pour les valeurs des anomalies
t['Avg_Anomaly_deg_C'] = None

# Réglez l'index du DataFrame sur la colonne de date (DateTime index)
t.set_index('date', inplace=True)

# Afficher les premiers éléments
t.head()
```

```
# Importer des bibliothèques pertinentes
import calendar
from datetime import datetime

# Définition de la fonction
def populate_df_with_anomalies_from_row(row):
    year = row['Year']
    # Anomaly values (elles semblent être un mélange de strings et de floats).
    monthly_anomalies = row.iloc[1:]
    # Noms abrégés des mois (noms d'index)
    months = monthly_anomalies.index
    for month in monthly_anomalies.index:
        # Obtenez le dernier jour de chaque mois 
        last_day = calendar.monthrange(year,datetime.strptime(month, '%b').month)[1]
        # construire l'index avec lequel nous pouvons référencer notre nouveau DataFrame 
        date_index = datetime.strptime(f'{year} {month} {last_day}', '%Y %b %d')
        # Remplir / régler la valeur @ au-dessus de l'index, à la valeur de l'anomalie
        t.loc[date_index] = monthly_anomalies[month]

# Appliquer la fonction à chaque ligne de données brutes 
_ = df_t.apply(lambda row: populate_df_with_anomalies_from_row(row), axis=1)

# Montrer les premiers éléments de notre DataFrame 
t.head()
```

```
# Définir la fonction pour convertir les valeurs en valeurs flottantes, et retourner un'NaN = Not a Number' si ce n'est pas possible.
def clean_anomaly_value(raw_value):
    try:
        return float(raw_value)
    except:
        return np.NaN
    
# Appliquer la fonction ci-dessus à toutes les valeurs de "anomaly" dans DataFrame
t['Avg_Anomaly_deg_C'] = t['Avg_Anomaly_deg_C'].apply(lambda raw_value: clean_anomaly_value(raw_value))

# 'Forward fill' pour prendre en compte les valeurs NaN
t.fillna(method='ffill', inplace=True)

# Montrer les premiers éléments de notre DataFrame récemment nettoyé
t.head()
```
